```python
import numpy as np
from sklearn.datasets import make_regression
from layer import Dense, RBFDense
from model import FFN
from functions import ReLU, Linear, Sigmoid
from losses import MSE, MAE, Huber
from optimisers import GeneticAlgorithmOptimizer, SGDOptimizer, ConjugateSGDOptimizer
```

```python
X, y = make_regression(n_samples=20000, n_features=15, n_informative=10)
y = (y-y.min())/(y.max()-y.min())
X = (X-X.min())/(X.max()-X.min())
```

```python
X_train, y_train = X[:10000], np.expand_dims(y[:10000], -1)
```

```python
X_test = X[10000:12000]
y_test = np.expand_dims(y[10000:12000], -1)
```

```python
model = FFN(input_shape = X_train.shape[1], layer_specs=[RBFDense(5), Dense(1, nonlinearity=Linear)], loss=MSE, reg_coef=0.1)
#model.compile(input_size = X_train.shape[1], output_size=1, layer_sizes=[25])
```

```python
trained, history = model.fit(train_sample=(X_train, y_train),
                   validation_sample=(X_test, y_test),
                   optimiser=GeneticAlgorithmOptimizer(35),
                   batch_size=32,
                   iters=150,
                   verbose=False)
```

```python
trained, history = model.fit(train_sample=(X_train, y_train),
                   validation_sample=(X_test, y_test),
                   optimiser=SGDOptimizer(batch_size=16, alpha=0.4, tol=1e-4),
                   batch_size=32,
                   iters=1000,
                   verbose=False)
```
