```python
import numpy as np
from sklearn.datasets import make_regression

from model import FeedForwardRegressor
from losses import MSE, MAE, Huber
from optimisers import GeneticAlgorithmOptimizer, NesterovMomentumOptimizer
```

```python
X, y = make_regression(n_samples=10000, n_features=15, n_informative=10)
y = (y-y.min())/(y.max()-y.min())
#X = (X-X.min())/(X.max()-X.min())
```

```python
X_train, y_train = X[:1000], y[:1000]
```

```python
X_test = X[1000:1200]
y_test = y[1000:1200]
```

```python
model = FeedForwardRegressor()
model.compile(input_size = X_train.shape[1], output_size=1, layer_sizes=[25])
```

```python
trained, history = model.fit(train_sample=(X_train, y_train),
                   validation_sample=(X_test, y_test),
                   loss=Huber, optimiser=GeneticAlgorithmOptimizer(35),
                   batch_size=16,
                   iters=10,
                   verbose=False)
```

```python
trained, history = model.fit(train_sample=(X_train, y_train),
                   validation_sample=(X_test, y_test),
                   loss=Huber, optimiser=NesterovMomentumOptimizer(alpha=0.25, batch_size=1),
                   batch_size=16,
                   iters=1000,
                   verbose=False)
```

```python
import seaborn as sns
sns.set()
```

```python
import pandas as pd
sns.lineplot(x='iteration', y='train_loss', data=pd.DataFrame(history))
sns.lineplot(x='iteration', y='validation_loss', data=pd.DataFrame(history))
```

```python
sns.lineplot(x='iteration', y='train_loss', data=pd.DataFrame(history))
sns.lineplot(x='iteration', y='validation_loss', data=pd.DataFrame(history))
```

```python
predicted = trained(X_test)
```

```python
sns.lineplot(x=np.linspace(0,1, len(predicted)), y=predicted.ravel())
sns.lineplot(x=np.linspace(0,1, len(y_test)), y=y_test)
```

```python
mse = MSE(predicted.ravel(), y_test)
```
