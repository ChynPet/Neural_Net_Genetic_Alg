```python
import numpy as np
from sklearn.datasets import make_regression

from model import FeedForwardRegressor
from functions import MeanSquared, Variable
from optimisers import GeneticAlgorithmOptimizer
```

```python
X, y = make_regression(n_samples=1000, n_features=15, n_informative=10)
y = (y-y.min())/(y.max()-y.min())
```

```python
X_train, y_train = X[:100], y[:100]
```

```python
model = FeedForwardRegressor([5, 15, 10, 15, 5], 100)
model.compile(input_size = X_train.shape[1], batch_size=X_train.shape[0], loss=MeanSquared, optimiser=GeneticAlgorithmOptimizer(35))
```

```python
trained= model.fit(X_train, y_train)
```

```python
x_var = Variable('x')
y_var = Variable('y')
```

```python
rel = MeanSquared(x_var, y_var)
```
