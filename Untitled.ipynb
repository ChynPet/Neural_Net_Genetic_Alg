```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_regression
from matplotlib import pyplot as plt

from layer import Dense, RBFDense
from model import FFN
from functions import ReLU, Linear, Sigmoid
from losses import MSE, MAE, Huber
from optimisers import GeneticAlgorithmOptimizer, SGDOptimizer, ConjugateSGDOptimizer

%matplotlib inline
```

```python
X, y = make_regression(n_samples=20000, n_features=15, n_informative=10)
y = (y-y.min())/(y.max()-y.min())
X = (X-X.min())/(X.max()-X.min())
```

```python
X_train, y_train = X[:10000], np.expand_dims(y[:10000], -1)
```

```python
X_test = X[10000:12000]
y_test = np.expand_dims(y[10000:12000], -1)
```

```python
model_sgd = FFN(input_shape = X_train.shape[1], layer_specs=[Dense(5, nonlinearity=ReLU), Dense(1, nonlinearity=Linear)], loss=MSE, reg_coef=0.1)
model_csgd = FFN(input_shape = X_train.shape[1], layer_specs=[Dense(5, nonlinearity=ReLU), Dense(1, nonlinearity=Linear)], loss=MSE, reg_coef=0.1)
```

```python
trained_sgd, history_sgd = model_sgd.fit(train_sample=(X_train, y_train),
                   validation_sample=(X_test, y_test),
                   optimiser=SGDOptimizer(batch_size=16),
                   batch_size=32,
                   iters=10_000,
                   verbose=False)
```

```python
trained_csgd, history_csgd = model_csgd.fit(train_sample=(X_train, y_train),
                   validation_sample=(X_test, y_test),
                   optimiser=ConjugateSGDOptimizer(batch_size=16),
                   batch_size=32,
                   iters=10_000,
                   verbose=False)
```

```python
plt.title('Train loss')

plt.plot(history_sgd['train_loss'], label='SGD')
plt.plot(history_csgd['train_loss'], label='CSGD')

plt.legend()
plt.show()

plt.title('Vall loss')

plt.plot(history_sgd['validation_loss'], label='SGD')
plt.plot(history_csgd['validation_loss'], label='CSGD')

plt.legend()
plt.show()
```

```python
predicted_sgd = trained_sgd.predict(X_test)
predicted_csgd = trained_csgd.predict(X_test)

print(f"SGD vall loss : {MSE(y_test, predicted_sgd)[0]}")
print(f"CSGD vall loss : {MSE(y_test, predicted_csgd)[0]}")
```

```python
plt.title('Prediction Results')

plt.plot(y_test[:,0], label='True')
plt.plot([np.median(y_test[:,0])]*y_test.shape[0], label='True Median')
plt.plot(predicted_csgd[:,0], label='CSGD')
plt.plot(predicted_sgd[:,0], label='SGD')

plt.legend()
plt.show()
```
